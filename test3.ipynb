```python
# %%
import numpy as np
import matplotlib.pyplot as plt

# %%
# -------------------------
# Define spherical bins
# -------------------------

v_edges = np.linspace(0, 5, 40)
theta_edges = np.linspace(0, np.pi, 30)
phi_edges = np.linspace(0, 2*np.pi, 32)

dv = v_edges[1] - v_edges[0]
dtheta = theta_edges[1] - theta_edges[0]
dphi = phi_edges[1] - phi_edges[0]

v_centers = 0.5*(v_edges[:-1] + v_edges[1:])
theta_centers = 0.5*(theta_edges[:-1] + theta_edges[1:])
phi_centers = 0.5*(phi_edges[:-1] + phi_edges[1:])

# %%
# -------------------------
# Build 3D mesh in spherical
# -------------------------

V, Theta, Phi = np.meshgrid(
    v_centers,
    theta_centers,
    phi_centers,
    indexing='ij'
)

# %%
# -------------------------
# Sample distribution
# Isotropic Maxwellian
# -------------------------

sigma = 1.0

f_vals = np.exp(-V**2 / (2*sigma**2))

# %%
# -------------------------
# Convert to Cartesian
# -------------------------

vx = V * np.sin(Theta) * np.cos(Phi)
vy = V * np.sin(Theta) * np.sin(Phi)
vz = V * np.cos(Theta)

# %%
# -------------------------
# Compute bin volume weights
# d^3v = v^2 sin(theta) dv dtheta dphi
# -------------------------

weights = (
    f_vals
    * V**2
    * np.sin(Theta)
    * dv * dtheta * dphi
)

# %%
# -------------------------
# Project onto vx-vy plane
# Integrate over vz
# -------------------------

vx_bins = np.linspace(-4, 4, 150)
vy_bins = np.linspace(-4, 4, 150)

F, vx_edges, vy_edges = np.histogram2d(
    vx.flatten(),
    vy.flatten(),
    bins=[vx_bins, vy_bins],
    weights=weights.flatten()
)

# %%
# -------------------------
# Plot
# -------------------------

plt.figure(figsize=(6,6))
plt.pcolormesh(vx_edges, vy_edges, F.T, shading='auto')
plt.gca().set_aspect('equal')
plt.xlabel(r"$v_x$")
plt.ylabel(r"$v_y$")
plt.title("Projected Maxwellian onto $v_x v_y$ plane")
plt.colorbar(label="Integrated Density")
plt.show()
```
